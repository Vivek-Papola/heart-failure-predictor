## Function of Sigmoid in Neural Network

The sigmoid function squashes the raw logits to a range between 0 and 1, making it suitable for binary classification tasks where you need a probability output

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    result = 1 / (1 + (np.exp(1) ** -z))
    return result
```

```python
sigmoid_res = [ sigmoid(i) for i in range(-20,20)]
```

```python
plt.plot([i for i in range(-20,20)], sigmoid_res ,marker='.')
plt.xlabel('Raw values')
plt.ylabel('Sigmoid Values')
plt.title('Plotting Sigmoid Function')
plt.grid()
plt.show()
```
